```python
#Removing dodgy images
data_dir='data'
img_exts=['jpeg','jpg','bmp','png']
for image_class in os.listdir(data_dir): 
    for image in os.listdir(os.path.join(data_dir, image_class)):
        image_path = os.path.join(data_dir, image_class, image)
        try: 
            img = cv2.imread(image_path)
            tip = imghdr.what(image_path)
            if tip not in img_exts: 
                print('Image not in ext list {}'.format(image_path))
                os.remove(image_path)
        except Exception as e: 
            print('Issue with image {}'.format(image_path))
            # os.remove(image_path)
```

```python
#Create the dataset
data=tf.keras.utils.image_dataset_from_directory('data') #Create dataset pipeline 
data_iterator=data.as_numpy_iterator() #allows us to access datapipeline by looping through the images 
#use as_numpy_iterator to inspect the content of our dataset
batch=data_iterator.next()  #it will pull some batch of data(32) from the pipeline 
```

## So label 0-happy and 1-sad

```python
#Standardizing the data
data=data.map(lambda x,y:(x/255,y))
```

```python
data.as_numpy_iterator().next()[0].max()
```

```python
len(data)
```

```python
#Splitting the data size into train,test,validation
train_size=int(len(data)*0.7)+1#19
validation_size=int(len(data)*0.2)+2#7
test_size=int(len(data)*0.1)-1#1
```

```python
test_size+train_size+validation_size
```

```python
#Actually taking the data 
train=data.take(train_size)
test=data.skip(train_size).take(test_size)
val=data.skip(train_size+test_size).take(validation_size)
```

```python
train
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Flatten,Conv2D,MaxPooling2D,GlobalMaxPooling2D,Dense,Input
```

```python
i=Input(shape=(256,256,3))
        
x=Conv2D(32,(3,3),activation='relu',padding='same')(i)
x=MaxPooling2D((2,2))(x)

x=Conv2D(64,(3,3),activation='relu',padding='same')(x)
x=MaxPooling2D((2,2))(x)

x=Conv2D(128,(3,3),activation='relu',padding='same')(x)
x=MaxPooling2D((2,2))(x)

x=Conv2D(128,(3,3),activation='relu',padding='same')(x)
x=MaxPooling2D((2,2))(x)

x=Flatten()(x)

x=Dense(512,activation='relu')(x)
x=Dense(1,activation='sigmoid')(x)

model=Model(i,x)
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
logdir='logs'
tensorboard_callback=tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
history=model.fit(train,validation_data=val,epochs=20,callbacks=[tensorboard_callback])
```

```python
losses=pd.DataFrame(history.history)
losses.plot()
```

```python
plt.plot(history.history['loss'],label='loss')
plt.plot(history.history['val_loss'],label='val_loss')
plt.legend()
plt.show()
```

```python
plt.plot(history.history['accuracy'],label='acc')
plt.plot(history.history['val_accuracy'],label='val_acc')
plt.legend()
```

```python
#instanciating the classification metrics
from tensorflow.keras.metrics import Precision,Recall,BinaryAccuracy
pre=Precision()
re=Recall()
acc=BinaryAccuracy()
```

```python
#Checking the accuracy,precision and recall
for batch in test.as_numpy_iterator():
    X,y=batch
    predictions=model.predict(X)
    pre.update_state(y,predictions)
    re.update_state(y,predictions)
    acc.update_state(y,predictions)
```

```python
print(f'Precision:{pre.result().numpy()},Recall:{re.result().numpy()},BinaryAccuracy:{acc.result().numpy()}')
```
